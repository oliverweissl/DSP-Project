```python
import matplotlib.pyplot as plt
```

## Plotting Loss Curves

```python
steps = [100*i + 1 for i in range(11)]

# Values extracted from openai api endpoint
flint_tr_loss = [0.5574,0.5189, 0.2297, 0.7785, 0.2060, 0.5905, 0.5329, 0.1526, 0.2807, 0.2433, 0.2147]
flint_val_loss = [1.1568, 0.5264, 0.2759, 0.2750, 0.1500, 0.5646, 0.5233, 0.2365, 0.6753, 0.3141, 0.4618]
```

```python
f, ax = plt.subplots()
ax.plot(steps, flint_tr_loss, color="b", label="Training Loss")
ax.plot(steps, flint_val_loss, color="g", label="Validation Loss")

ax.set_ylabel("Loss")
ax.set_xlabel("Steps")
ax.set_ylim([0.1, 1.4])
plt.legend()

plt.savefig("figures/flint_loss_curve.png", dpi=200)
```
